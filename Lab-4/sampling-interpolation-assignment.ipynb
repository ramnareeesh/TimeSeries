# Dataset: `Month_Value_1.csv`

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def parser(x):
    return pd.to_datetime(x, format='%d.%m.%Y')
```

```python
dataset = pd.read_csv("../data/Month_Value_1.csv", parse_dates=["Period"], date_format="%d.%m.%Y")
dataset["Period"] = dataset["Period"].apply(parser)
dataset.dropna(axis=0, inplace=True)
dataset.head(35)
```

## Manual plotting without setting index column

```python
plt.plot(dataset["Period"], dataset["Sales_quantity"])
plt.title("Sales Quantity")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(dataset["Period"][::10], rotation=45)
plt.show()
```

## After setting index column as "Month"

```python
dataset.set_index('Period', inplace=True)
dataset.head(40)
```

```python
dataset["Sales_quantity"].plot()
plt.title("Sales Quantity")
plt.show()
```

# Upsampling

## Upsampling to daily sales information

```python
upsampled = dataset.resample('D').mean()
upsampled.head(35)
```

# Interpolation

## Linear Interploation

```python
interpolated_linear = upsampled.interpolate(method="linear")
interpolated_linear
```

```python
interpolated_linear["Sales_quantity"].plot()
plt.title("Interpolated [Linear] Sales Quantity with Upsampling")
plt.show()
```

## Spline Interpolation

```python
interpolated_spline = upsampled.interpolate(method="spline", order=2)
interpolated_spline
```

```python
interpolated_spline["Sales_quantity"].plot()
plt.title("Interpolated [Spline(2)] Sales Quantity with Upsampling")
plt.show()
```

# Downsampling

## Downsampling to quarterly sales information

```python
downsampled = dataset.resample('QE').mean()
downsampled
```

```python
downsampled["Sales_quantity"].plot()
plt.title("Quarterly sales quantity information")
plt.show()
```

## Downsampling to yearly sales information

```python
downsampled_2 = dataset.resample('YE').mean()
downsampled_2
```

```python
downsampled_2["Sales_quantity"].plot()
plt.title("Yearly shampoo sales information")
plt.show()
```
